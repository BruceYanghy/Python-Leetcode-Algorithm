String t is generated by random shuffling string s and then add one more letter at a random position.

Example:

Input:
s = "abcd"
t = "abcde"

Output:
e

Explanation:
'e' is the letter that was added.

```python
def findTheDifference(s, t):
        """
        :type s: str
        :type t: str
        :rtype: str
        """
        letters = [0] * 26
        for c in s:
            letters[ord(c) - 97] += 1
        for c in t:
            letters[ord(c) - 97] -= 1
            if letters[ord(c) - 97] < 0:
                return c
```

```python
def findTheDifference(s, t):
        """
        :type s: str
        :type t: str
        :rtype: str
        """
        letters = {}
        for c in s:
            letters[c] = letters[c] + 1 if c in letters else 1
        for c in t:
            if c not in letters:
                return c
            letters[c] -= 1
            if letters[c] < 0:
                return c
```

```python
class Solution(object):
    def findTheDifference(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: str
        """
        res = ord(t[-1])
        for i in xrange(len(s)):
            res = res + ord(t[i]) - ord(s[i])
        return chr(res)
```

```python
def findTheDifference(self, s, t):
        dic = {}
        for ch in s:
            dic[ch] = dic.get(ch, 0) + 1
        for ch in t:
            if dic.get(ch, 0) == 0:
                return ch
            else:
                dic[ch] -= 1
```

# 思路：考虑到t是由s添加一个字母得到，那么这个添加的字母在s和t中出现的总次数一定是唯一的一个奇数。基于这个想法，用异或操作得到该字母。

```python
s="abcdee"
t="abcdeea"
code = 0
for ch in s + t:
    code ^= ord(ch)
chr(code)
```

```python
class Solution(object):
    """
    xor
    """
    def findTheDifference(self, s, t):
        code = 0
        for ch in s + t:
            code ^= ord(ch)
        return chr(code)
```
